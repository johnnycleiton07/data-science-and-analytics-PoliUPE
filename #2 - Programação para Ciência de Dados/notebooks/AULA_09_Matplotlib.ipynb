## Recriando o gráfico da sigmoid com o pyplot

```python
def sigmoid(array):
  resultado = 1 / (1 + np.exp(-array))
  return resultado
```

```python
entrada = np.linspace(-8, 8, 100) # cria um array com 100 valores igualmente espaçados entre -8 e 8
entrada
```

```python
saida = sigmoid(entrada) # gerando a saída do gráfico
saida
```

image.png

```python
fig, ax = plt.subplots()
ax.plot(entrada, saida, color="red")
ax.axvline(0, color="black") # eixos y em 0
ax.axhline(0, color="black") # eixos x em 0
ax.grid(axis='y', alpha=0.4, linestyle='--') # grades
ax.grid(axis='x', alpha=0.4, linestyle='--')
```

O `pyplot` possui vários parâmetros

```python
fig, ax = plt.subplots()
ax.plot(entrada, saida, "r--")
ax.axvline(0, color="black") # eixos y em 0
ax.axhline(0, color="black") # eixos x em 0
ax.grid(axis='y', alpha=0.3, linestyle='--') # grades
ax.grid(axis='x', alpha=0.3, linestyle='--')
ax.set_xlabel("x")
ax.set_ylabel("sigmoid(x)")
```

```python
fig, axs = plt.subplots(nrows = 2, ncols = 1)
axs[0].plot(entrada, saida, "r--")
axs[0].axvline(0, color="black") # eixos y em 0
axs[0].axhline(0, color="black") # eixos x em 0
axs[0].grid(axis='y', alpha=0.3, linestyle='--') # grades
axs[0].grid(axis='x', alpha=0.3, linestyle='--')
# axs[0].set_xlabel("x")
axs[0].set_ylabel("sigmoid(x)")


saida_cos = np.cos(entrada)
axs[1].plot(entrada, saida_cos, "bo")
axs[1].axvline(0, color="black") # eixos y em 0
axs[1].axhline(0, color="black") # eixos x em 0
axs[1].grid(axis='y', alpha=0.3, linestyle='--') # grades
axs[1].grid(axis='x', alpha=0.3, linestyle='--')
axs[1].set_xlabel("x")
axs[1].set_ylabel("cos(x)")
```

Adicionando múltiplas curvas

- Basta plotar as novas curvas

```python
fig = plt.figure()
perc_esq = 0.1
perc_topo = 0.2
perc_largura = 0.8
perc_altura = 0.6

ax = fig.add_axes([perc_esq, perc_topo, perc_largura, perc_altura])
x = np.arange(0, 5, 0.1)
ax.set_xlabel("x")
ax.set_ylabel("f(x)")
ax.plot(x, y, label="Seno")
ax.set_title("Curva do Seno e Cosseno")

ax.plot(x, np.cos(x), label="Cosseno")
ax.legend()
```

# Personalização do gráfico
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot

```python
x = np.arange(0, 10, 0.1)
fig = plt.figure(figsize=(6,6))
ax = fig.add_axes([perc_esq, perc_topo, perc_largura, perc_altura])

ax.set_xlabel("x")
ax.set_ylabel("f(x)")

ax.plot(x, x**2, color="black", label="x^2", linestyle="-.")
ax.plot(x, 2**x, color="green", label="2^x", linewidth="2", alpha=0.5)
ax.legend()

```

## Limites das visualizações

```python
ax.set_xlim([0,8])
ax.set_ylim([0,60])
fig
```

```python
fig = plt.figure()
esquerda = 0.1
topo = 0.2
largura = 0.7
altura = 0.6
ax = fig.add_axes([esquerda, topo, largura, altura])
ax.set_title('Gráfico')
ax.plot(x, x**2, color="black", label="x^2", linestyle="-.")
ax.plot(x, 2**x, color="green", label="2^x", linewidth="2", alpha=0.5)
ax.legend(loc='upper right')


ax_zoom = fig.add_axes([0.2, 0.4, 0.3, 0.3])
ax_zoom.set_title('Zoom')
ax_zoom.set_xlim(0,8)
ax_zoom.set_ylim(0,60)
ax_zoom.plot(x, x**2, linestyle="-.", label="x^2")
ax_zoom.plot(x, 2**x, linewidth="2", label="2^x")
```

## Outros tipos de gráficos

### Scatter

```python
x = np.arange(0, 10, 0.1)
fig = plt.figure(figsize = (6, 6))
ax = fig.add_axes([perc_esq, perc_topo, perc_largura, perc_altura])

ax.set_xlabel("x")
ax.set_ylabel("f(x)")
ax.scatter(x, x**2, color="black", label="x^2")
ax.scatter(x, 2**x, color="green", label="2^x")

ax.set_xlim([0, 5])
ax.set_ylim([0, 30])
ax.legend()
```
