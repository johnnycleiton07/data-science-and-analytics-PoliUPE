Podemos usar uma estrutura de repetição *while* para percorrer as posições de uma lista.


image.png

````python
# O que o código anterior faz?
notas = [6, 7, 5, 8 , 9]
soma = 0
x = 0
while x < 5:
  soma = soma + notas[x]
  x = x + 1
print(f"Média: {soma/5:.2f}")
````

# Tuplas

Tuplas podem ser vistas como listas em Python, com a grande diferença de serem imutáveis.

Elas são ideias para representar listas de valores constantes e para realizar operações nas funções.

````python
# Definindo uma tupla
tupla = ("a", "b", "c")
print(tupla)
````

````python
print(type(tupla))
````

````python
# Acessando a uma posição específica
print(tupla[0])
````

````python
# Fatiamento de uma tupla
print(tupla[1:])
````

````python
# Repetindo uma tupla0
print(tupla * 2)
````

````python
# Obtendo o tamanho da tupla
print(len(tupla))
````

````python
# Essa operação é possível?
tupla[0] = "A"
````

Podemos transformar uma tupla em uma lista com a função `tuple`

````python
lista = [1, 2 , 3]
tupla = tuple(lista)
print(tupla)
````

````python
tupla = (1, 2 , 3)
lista = list(tupla)
print(lista)
````

# Dicionário

Dicionários (*dict*) consistem em uma estrutura de dados similar a lista, mas com propriedades de acesso diferentes.

Um dicionário é composto por um conjunto de chaves e valores. O dicionário em si consiste em relacionar uma chave a um valor específico.

Em Python, dicionários são criados com chaves ( { } ).



Cada elemento no dicionário é a combinação de chave/valor

# Estrutura de Repetição `for`

Python apresenta uma estrutura de repetição especialmente projetada para percorrer listas.

A estrutura de repetição for funciona de forma parecida a while, mas a cada repetição utiliza um elemento diferente da lista.

Em outras linguagens a estrutura for do python é chamada de forEach.


````python
lista = [8, 9, 15]
for valor in lista:
  print(valor)
````

equivalente a

````python
cont = 0
lista = [8, 9, 15]
while cont < len(lista):
  valor = lista[cont]
  print(valor)
  cont = cont + 1 # ou cont += 1
````

Embora a estrutura de repetição for facilite o nosso trabalho, ela não substitui totalmente a instrução while.

Normalmente, utilizamos for quando queremos percorrer uma variável que é iterável (lista, tupla, dicionário... ).

O while normalmente é utilizado quando não sabemos a quantidade de iterações (repetições).  

As instruções *break* e *continue* funcionam igualmente no `for`


````python
for num in [2, 4, 5, 6, 10]:
  if num % 2 != 0:
    print(num)
    break
````

# Exercícios

Vamos criar uma agenda de contatos telefônicos?
As principais operações são: adicionar, remover, editar e lista números.

````python
contatos = {}
````

````python
menu="""
1 - Adicionar contato
2 - Remover contato
3 - Editar contato
4 - Lista todos os contatos
5 - Sair
Opção escolhida: """
````

````python
while True:
  op = input(menu)

  if op == "1":
    print("Informações do contato: ")
    nome = input("Nome: ")

    if nome not in contatos:
      numero = input("Número: ")
      contatos[nome] = numero
    else:
      print(f"{nome} já foi inserido. Use a opção de editar (op=3).")

  elif op == "2":
    print("Iniciando remoção: ")
    nome = input("Nome")

    if nome in contatos:
      del contatos[nome]
      print(f"{nome} removido com sucesso.")
    else:
      print(f"{nome} não existe na agenda.")

  elif op == "3":
    print("Iniciando edição: ")
    nome = input("Nome: ")

    if nome in contatos:
      numero = input("Número: ")
      contatos[nome] = numero
    else:
      print(f"{nome} não existe. Use a opção de adicionar (op=1).")

  elif op == "4":

    print("Listando agenda...")
    for cont, nome in enumerate(contatos):
      numero = contatos[nome]
      print(f"{cont} - {nome} | {numero}")

  elif op == "5":
    print("Saindo da agenda...")
    break
  else:
    print("Opção inválida")

print(contatos)
````
