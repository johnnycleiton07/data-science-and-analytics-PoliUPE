````python
def softmax(mat):
  res = np.zeros(mat.shape)
  for i in range(mat.shape[0]):
    sume_x = np.sum(np.exp(mat[i]))
    res[i] = np.exp(mat[i])/sume_x
  return res
````

### Dot Product

image.png

````python
a = np.random.randn(100000)
b = np.random.randn(100000)
````

````python
# metodo tradicional
import time

inicio = time.process_time()
r = 0
for i in range(len(a)):
  r += a[i] * b[i]
final = time.process_time()

print(f"resultado: {r} - tempo: {(final-inicio)*1000:.2f}ms")
````

````python
# usando numpy
inicio = time.process_time()
r = np.dot(a, b)
final = time.process_time()

print(f"resultado: {r} - tempo: {(final-inicio)*1000:.2f}ms")
````

### Outer Product

image.png

````python
a = np.random.randn(1000)
b = np.random.randn(1000)
````

````python
# metodo tradicional com loop
import time

inicio = time.process_time()
r = np.zeros((a.size, b.size))
for i in range(a.size):
  for j in range(b.size):
    r[i,j] = a[i] * b[j]
final = time.process_time()

print(f"resultado: {str(np.sum(r))} - tempo: {(final-inicio)*1000:.2f}ms")
````

````python
# com numpy
inicio = time.process_time()
r = np.outer(a, b)
final = time.process_time()

print(f"resultado: {str(np.sum(r))} - tempo: {(final-inicio)*1000}ms")
````

## Criando um novo tipo de dados com orientação a objetos

````python
from datetime import date

class Pessoa():
  def __init__(self, nome, data_nascimento): # construtor
    self._nome = nome
    self._data_nascimento = data_nascimento
    self._conjuge = None

  @property
  def nome(self):
    return self._nome

  @property
  def idade(self):
    diferenca = date.today() - self._data_nascimento
    return int(diferenca.days // 365.25)

  @property
  def conjuge(self):
    return self._conjuge

  @property
  def possui_conjuge(self):
    return self._conjuge != None

  @nome.setter
  def nome(self, novo_nome):
    if novo_nome:
      self._nome = novo_nome
    else:
      print("Você precisa informar um nome")

  def __configurar_conjuge(self, pessoa):
    if isinstance(pessoa, Pessoa):
      self._conjuge = pessoa
      print(f"{self.nome} iniciou um relacionamento com {pessoa.nome}")
    else:
      print("Erro ao adicionar um conjuge")

  def iniciar_relacionamento(self, pessoa):
    """
    Inicia um relacionamento entre o objeto atual e a `pessoa` passada por parâmetro.
    """
    self.__configurar_conjuge(pessoa)
    pessoa.__configurar_conjuge(self)

  def __repr__(self):
    """
    Imprimi os dados da pessoa
    """
    descricao = f"Nome={self._nome}, nascido em={self._data_nascimento}, "
    if not self.possui_conjuge:
      descricao += f"casado=não"
    else:
      descricao += f"cônjuge={self.conjuge.nome}"

    return descricao
````

````python
class Professor(Pessoa):
  def __init__(self, nome, data_nascimento, curso):
    super().__init__(nome, data_nascimento)
    self._disciplina = curso

  @property
  def disciplina(self):
    return self._disciplina
````
